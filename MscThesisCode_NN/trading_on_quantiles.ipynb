```python
def calculate_profits(corrected_ensembles, original_ensembles, trade_q, actuals, day_ahead_spot_prices_DK1, imbalance_prices_DK1, printing=False):
    profits_corrected = []
    profits_original = []
    amounts_corrected = []
    amounts_original = []

    for hour_ in corrected_ensembles.index:
        buy_amount_corrected = corrected_ensembles.loc[hour_].iloc[int(trade_q * len(corrected_ensembles.loc[hour_]))]
        buy_amount_original = original_ensembles.loc[hour_].sort_values().iloc[int(trade_q * len(original_ensembles.loc[hour_]))]

        if printing:
            print("buy_amount (corrected and original)", buy_amount_corrected, ";", buy_amount_original)

        actuals_obs = actuals.loc[hour_]

        profit_corrected = day_ahead_spot_prices_DK1["SpotPriceDKK"].loc[hour_] * buy_amount_corrected
        energy_to_sell_at_imbalance_corrected = actuals_obs - buy_amount_corrected
        profit_corrected += imbalance_prices_DK1["ImbalancePriceDKK"].loc[hour_] * energy_to_sell_at_imbalance_corrected

        profit_original = day_ahead_spot_prices_DK1["SpotPriceDKK"].loc[hour_] * buy_amount_original
        energy_to_sell_at_imbalance_original = actuals_obs - buy_amount_original
        profit_original += imbalance_prices_DK1["ImbalancePriceDKK"].loc[hour_] * energy_to_sell_at_imbalance_original

        if printing:
            print(f"The profit from the corrected ensembles for hour {hour_} is: {float(profit_corrected.iloc[0]):.2f}, \nFrom original ensembles is: {float(profit_original.iloc[0]):.2f}")

        profits_corrected.append(float(profit_corrected.iloc[0]))
        profits_original.append(float(profit_original.iloc[0]))
        amounts_corrected.append(buy_amount_corrected)
        amounts_original.append(buy_amount_original)

    return profits_corrected, profits_original, amounts_corrected, amounts_original

profits_corrected, profits_original, amounts_corrected, amounts_original = calculate_profits(corrected_ensembles, original_ensembles, trade_q, actuals, day_ahead_spot_prices_DK1, imbalance_prices_DK1, printing=False)
```

```python
def calculate_profits_sde(corrected_ensembles, original_ensembles, SDE_corrected_ensembles, trade_q, actuals, day_ahead_spot_prices_DK1, imbalance_prices_DK1, printing=False):
    profits_corrected = []
    profits_original = []
    profits_sde = []
    amounts_corrected = []
    amounts_original = []
    amounts_sde = []

    for hour_ in corrected_ensembles.index:
        buy_amount_corrected = corrected_ensembles.loc[hour_].iloc[int(trade_q * len(corrected_ensembles.loc[hour_]))]
        buy_amount_original = original_ensembles.loc[hour_].sort_values().iloc[int(trade_q * len(original_ensembles.loc[hour_]))]

        if printing:
            print("buy_amount (corrected and original)", buy_amount_corrected, ";", buy_amount_original)

        actuals_obs = actuals.loc[hour_]

        profit_corrected = day_ahead_spot_prices_DK1["SpotPriceDKK"].loc[hour_] * buy_amount_corrected
        energy_to_sell_at_imbalance_corrected = actuals_obs - buy_amount_corrected
        profit_corrected += imbalance_prices_DK1["ImbalancePriceDKK"].loc[hour_] * energy_to_sell_at_imbalance_corrected

        profit_original = day_ahead_spot_prices_DK1["SpotPriceDKK"].loc[hour_] * buy_amount_original
        energy_to_sell_at_imbalance_original = actuals_obs - buy_amount_original
        profit_original += imbalance_prices_DK1["ImbalancePriceDKK"].loc[hour_] * energy_to_sell_at_imbalance_original

        if printing:
            print(f"The profit from the corrected ensembles for hour {hour_} is: {float(profit_corrected.iloc[0]):.2f}, \nFrom original ensembles is: {float(profit_original.iloc[0]):.2f}")

        # print(profit_corrected , "<- profit_corrected")
        profits_corrected.append(float(profit_corrected))
        profits_original.append(float(profit_original))
        amounts_corrected.append(buy_amount_corrected)
        amounts_original.append(buy_amount_original)

    for hour_ in SDE_corrected_ensembles.index:
        if hour_ in actuals.index:
            buy_amount_sde = SDE_corrected_ensembles.loc[hour_].iloc[int(trade_q * len(SDE_corrected_ensembles.loc[hour_]))]

            if printing:
                print("buy_amount (SDE corrected)", buy_amount_sde)

            actuals_obs = actuals.loc[hour_]

            profit_sde = day_ahead_spot_prices_DK1["SpotPriceDKK"].loc[hour_] * buy_amount_sde
            energy_to_sell_at_imbalance_sde = actuals_obs - buy_amount_sde
            profit_sde += imbalance_prices_DK1["ImbalancePriceDKK"].loc[hour_] * energy_to_sell_at_imbalance_sde

            if printing:
                print(f"The profit from the SDE corrected ensembles for hour {hour_} is: {float(profit_sde):.2f}")

            profits_sde.append(float(profit_sde))
            amounts_sde.append(buy_amount_sde)

    return profits_corrected, profits_original, profits_sde, amounts_corrected, amounts_original, amounts_sde

# Example usage
```

```python
quantiles = np.linspace(0.05, 0.95, 20)

def profits_for_all_quantiles(corrected_ensembles, original_ensembles, actuals, day_ahead_spot_prices_DK1, imbalance_prices_DK1, printing=False, quantiles_ = quantiles):
    profits_corrected = {}
    profits_original = {}
    amounts_corrected = {}
    amounts_original = {}

    for trade_q in quantiles_:
        profits_corrected[trade_q], profits_original[trade_q], amounts_corrected[trade_q], amounts_original[trade_q] = calculate_profits(corrected_ensembles, original_ensembles, trade_q, actuals, day_ahead_spot_prices_DK1, imbalance_prices_DK1, printing=printing)

    return profits_corrected, profits_original, amounts_corrected, amounts_original

profits_corrected, profits_original, amounts_corrected, amounts_original = profits_for_all_quantiles(corrected_ensembles, original_ensembles, actuals, day_ahead_spot_prices_DK1, imbalance_prices_DK1, printing=False, quantiles_=quantiles)

#plot the mean profit for each quantile in one plot
mean_profits_corrected = [np.mean(profits_corrected[trade_q]) for trade_q in quantiles]
mean_profits_original = [np.mean(profits_original[trade_q]) for trade_q in quantiles]
```

```python
import numpy as np
import matplotlib.pyplot as plt

# Example dictionaries with keys and profits
keys = profits_corrected.keys()  # Assuming keys are the same for both dictionaries

# Set up the figure
fig, ax = plt.subplots(figsize=(15, 8))

# Create a flipped histogram for each key
max_hist_value_corrected = 0  # To find the maximum value of histograms for normalization in corrected data
max_hist_value_original = 0  # To find the maximum value of histograms for normalization in original data

# Compute histograms and find the maximum values for normalization
hist_data_corrected = []
hist_data_original = []
mean_values = []

for key in keys:
    profits_corrected_values = profits_corrected[key]
    profits_original_values = profits_original[key]
    
    hist_corrected, bins_corrected = np.histogram(profits_corrected_values, bins=30, density=True)
    hist_original, bins_original = np.histogram(profits_original_values, bins=30, density=True)
    
    hist_data_corrected.append((hist_corrected, bins_corrected))
    hist_data_original.append((hist_original, bins_original))
    
    max_hist_value_corrected = max(max_hist_value_corrected, hist_corrected.max())
    max_hist_value_original = max(max_hist_value_original, hist_original.max())
    
    mean_values.append(np.mean(profits_corrected_values))

# Plot histograms
for i, ((hist_corrected, bins_corrected), (hist_original, bins_original)) in enumerate(zip(hist_data_corrected, hist_data_original)):
    # Normalize histograms to fit within the x-axis range
    normalized_hist_corrected = hist_corrected / max_hist_value_corrected
    normalized_hist_original = hist_original / max_hist_value_original
    
    # Compute bin centers
    bin_centers_corrected = 0.5 * (bins_corrected[1:] + bins_corrected[:-1])
    bin_centers_original = 0.5 * (bins_original[1:] + bins_original[:-1])
    
    # Plot the original histogram as a light blue bar plot, flipped horizontally and shifted to the key position
    ax.barh(bin_centers_original, normalized_hist_original, height=(bins_original[1] - bins_original[0]), left=i, color='red', alpha=0.2, edgecolor='red')
    
    # Plot the corrected histogram as a black bar plot, flipped horizontally and shifted to the key position
    ax.barh(bin_centers_corrected, normalized_hist_corrected, height=(bins_corrected[1] - bins_corrected[0]), left=i, color='black', alpha=0.3, edgecolor='black')
    
    # Plot the mean value as a horizontal line
    # ax.plot([i, i + normalized_hist_corrected.max()], [mean_values[i], mean_values[i]], color='red', linestyle='--', linewidth=1)

# Set x-ticks as keys
ax.set_xticks(range(len(keys)))
ax.set_xticklabels([f'{key:.2f}' for key in keys], rotation=90)

# Labels and title
ax.set_xlabel('Quantiles')
ax.set_ylabel('Profit Distribution')
ax.set_title('Flipped Histograms of Profits')

# Custom legend
handles = [
    plt.Line2D([0], [0], color='red', lw=4, label='Original Profits'),
    plt.Line2D([0], [0], color='black', lw=4, label='Corrected Profits'),
]
ax.legend(handles=handles, loc='upper right')

# Show plot
plt.tight_layout()
# plt.savefig("Flipped_Histograms_of_Profits.pdf")
plt.show()
```
